```python
# Write your code here
# Q-1.1-A
import pandas as pd
import urllib.request
dataurl = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
data = urllib.request.urlopen(dataurl)
rawdata = []
for line in data:
  changed_data = line.decode("utf-8").replace('\r\n', '')
  if changed_data: rawdata.append(changed_data)

modified_data = pd.DataFrame(rawdata, columns=['Raw Data'])

modified_data['Sentence Count']=modified_data['Raw Data'].apply(lambda x: len(str(x).split(".")))
modified_data
```

```python
#Q-1.1-B
modified_data['Word Count']=modified_data['Raw Data'].apply(lambda x: len(str(x).split(" ")))
modified_data
```

```python
#Q-1.1-C
modified_data['Character Count']=modified_data['Raw Data'].str.len()
modified_data
```

```python
#Q-1.1-D
def avg_word(sentence):
  words = sentence.split()
  if len(words) is not 0:
    return (sum(len(word) for word in words)/len(words))

modified_data['Average Word Count'] = modified_data['Raw Data'].apply(lambda x: avg_word(x))
modified_data
```

```python
#Q-1.1-E
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')

modified_data['Stop Words Count'] = modified_data['Raw Data'].apply(lambda x: len([x for x in x.split() if x in stop]))
modified_data
```

```python
#Q-1.1-F
import string
special_characters = string.punctuation

def findspecial_characters(sentence):
  special_char = 0
  for i in range(0, len(sentence)):
    ch = sentence[i]
    if(sentence[i].isalpha()):
      continue
    elif(sentence[i].isdigit()):
      continue
    else:
      special_char += 1
    
  return (special_char)

modified_data['Special Character Count'] = modified_data['Raw Data'].apply(lambda x: findspecial_characters(x))
modified_data
```

```python
#Q-1.1-G
modified_data['Numerics Count'] = modified_data['Raw Data'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
modified_data
```

```python
#Q-1.1-H
modified_data['UpperCase Count'] = modified_data['Raw Data'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
modified_data
```

```python
#Q-1.2-A
modified_data['LowerCase'] = modified_data['Raw Data'].apply(lambda x: " ".join(x.lower() for x in x.split()))
modified_data
```

```python
#Q-1.2-B
import string
modified_data['Removed Punctuation'] = modified_data['Raw Data'].apply(lambda x: "".join(x.lower() for x in x if x not in string.punctuation))
modified_data
```

```python
#Q-1.2-C
from nltk.corpus import stopwords
stop = stopwords.words('english')
modified_data['Removed StopWords'] = modified_data['Raw Data'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
modified_data
```

```python
#Q-1.2-D
frequent_words = pd.Series(' '.join(modified_data['Raw Data']).split()).value_counts()[:10]
print(frequent_words)
frequent_words = list(frequent_words.index)
modified_data['Removed CommonWords'] = modified_data['Raw Data'].apply(lambda x: " ".join(x for x in x.split() if x not in frequent_words))
modified_data
```

```python
#Q-1.2-E
rare_words = pd.Series(' '.join(modified_data['Raw Data']).split()).value_counts()[-10:]
print(rare_words)
rare_words = list(rare_words.index)
modified_data['Removed RareWords'] = modified_data['Raw Data'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_words))
modified_data
```

```python
#Q-1.2-F
from textblob import TextBlob
modified_data['Spelling Correction'] = modified_data['Raw Data'].apply(lambda x: str(TextBlob(x).correct()))
modified_data
```

```python
#Q-1.2-G
import nltk
nltk.download('punkt')
modified_data['Tokenization'] = modified_data['Raw Data'].apply(lambda x: TextBlob(x).words)
modified_data
```

```python
#Q-1.2-H
from nltk.stem import PorterStemmer
st = PorterStemmer()
modified_data['Stemming'] = modified_data['Raw Data'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
modified_data
```

```python
#Q-1.2-I
from textblob import Word
import nltk
nltk.download('wordnet')
modified_data['Lemmatization'] = modified_data['Raw Data'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
modified_data
```

```python
# Q-1.3
import csv
modified_data.to_csv("Cleaned_Sentences.csv", index= False)
```

```python
# Q-1.4-A
text_frequency = (modified_data['Raw Data']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
text_frequency.columns = ['words','frequency']
text_frequency
```

## 2.1 Write a Python program to remove leading zeros from an IP address. (4 points)

ip = "260.08.094.109"

```python
# Write your code here
import re
ip = "260.08.094.109"
splittingip = ip.split(".")
output = []
for i in splittingip:
  x = re.sub("^0","",i)
  output.append(x)
newip = ".".join(output)
print(newip)
```
