```python
#define variables
status = []
name = []
summary = []
jurisdiction = []
api = "https://api.case.law/v1/cases/"
```

```python
# collect data from 100 cases
import requests
for caseID in range(236600,236700):
    url = api + str(caseID) + "/?full_case=true"
    response = requests.get(url, headers={'Authorization': 'Token 580051d449440d3301abf6f62ab3e580c792dcc9'})
    case = response.json()
    status.append(case['casebody']['status'])
    name.append(case['name'])
    jurisdiction.append(case['jurisdiction']['name_long'])
    case_body = case['casebody']['data']['opinions']
    summary.append(case_body[0]['text'])
        
```

```python
import pandas as pd
df = pd.DataFrame({'Case Name': name,'Status': status,'Jurisdiction': jurisdiction, 'Summary': summary})
df
```

## Data Preprocessing

#### Lowercase

```python
df['LowerCase'] = df['Summary'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

#### Removed Punctuation

```python
import string
df['Removed Punctuation'] = df['LowerCase'].apply(lambda x: "".join(x.lower() for x in x if x not in string.punctuation))
df
```

#### Removed Noise

```python
df['Removed Noise'] = df['Removed Punctuation'].apply(lambda x: " ".join(x for x in x.split() if x.isalnum()))
df
```

```python
import nltk
nltk.download()
```

#### Removed Stopwords

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Removed StopWords'] = df['Removed Noise'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

#### Removing Frequent Words

```python
frequent_words = list(pd.Series(' '.join(df['Summary']).split()).value_counts()[:15].index)
df["Frequent Words"] = df['Removed StopWords'].apply(lambda x: ' '.join([i for i in x.split() if i not in frequent_words]))
df
```

#### Stemming

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['Stemming'] = df['Frequent Words'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
df
```

#### Word Tokenize

```python
from textblob import TextBlob
df['Word Tokenize'] = df['Frequent Words'].apply(lambda x: TextBlob(x).words)
df
```

#### Lemmatization

```python
from textblob import Word
import nltk
nltk.download('wordnet')
df['Lemmatization'] = df['Word Tokenize'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x]))
df
```

```python
cleaned_data = df[['Case Name','Lemmatization']]
```
